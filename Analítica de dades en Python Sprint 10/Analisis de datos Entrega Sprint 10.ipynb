Crea una columna nova que sigui la data de naixament en format Datetime a partir de les columnes dia, mes i any. Utilitzant aquesta columna crea una funció que donada una data, et calculi l'edat actual a dia d'avui.

Utilitza la funció que acabes de crear per generar una columna nova al DataFrame amb l'edat actual.

```python
#importamos la libreria y el objeto datetime
from datetime import datetime

#  Crear la columna Datetime uniendo Any, Mes y Dia, deben estar en ingles para que pandas las transforme
df['Fecha_Nacimiento'] = pd.to_datetime(
    df[['Año', 'Mes', 'Dia']].rename(columns={'Año': 'year', 'Mes': 'month', 'Dia': 'day'})
)

# Creamos la función  que calcula la edad exacta
def calcular_edad(fecha_nacimiento):
    hoy = datetime.today()
    edad = hoy.year - fecha_nacimiento.year
    
    # Restamos 1 si la fecha de hoy es anterior al día del cumpleaños este año
    if (hoy.month, hoy.day) < (fecha_nacimiento.month, fecha_nacimiento.day):
        edad -= 1
    return edad

# Aplicamos la función para rellenar la nueva columna de edad actual
df['Edad_Actual'] = df['Fecha_Nacimiento'].apply(calcular_edad)

# Mostramos las columnas de interés para comprobar que se hizo bien
df[['Fecha_Nacimiento', 'Edad_Actual']].head()
```

## 1.
Crea una funció que prengui un dataframe com a paràmetre d'entrada.

La funció ha de crear (i exportar) un gràfic automàticament per a cada columna del dataframe. Per exemple:

un histograma/boxplot si la variable és numèrica
unes barres dels valors més freqüents si és categòrica
unes barres dels anys més freqüents si la dada està en format data.
La idea és crear una funció que funcioni per qualsevol dataframe, no només amb el que hem treballat fins ara.

Mostra el resultat de la funció en algun dels datasets d’exemple que conté el paquet seaborn. Per exemple, iris, penguins o titanic.

Tingues en consideració que en el següent sprint treballaràs exclusivament amb gràfics. L’objectiu d’aquest exercici no és crear gràfics molt elaborats, sinó resoldre una necessitat de manera ràpida i automàtica.

```python
#Importamos las librerias necesarias para hacer el ejercicio
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


def generar_graficos_automaticos(dataframe):
    # Creamos la carpeta donde guardaremos todas las imagenes png
    os.makedirs("mis_graficos", exist_ok=True)
    
    for columna in dataframe.columns:
       #Quitamos los nulos de la columna antes de analizarla
        datos_limpios = dataframe[columna].dropna()
        
        # Si la columna se quedó vacía por culpa de los nulos, la saltamos
        if datos_limpios.empty:
            print(f"Saltando '{columna}': columna vacía o llena de nulos.")
            continue
            
        # Usamos try y except para evitar que el código se detenga por errores
        try:
            plt.figure(figsize=(7, 4))  #asignamos un numero para el tamaño de las imagenes
            
            # Condición para columnas numéricas
            if pd.api.types.is_numeric_dtype(datos_limpios):
                sns.histplot(x=datos_limpios, color="royalblue", kde=True)
                plt.title(f"Análisis Numérico: {columna}")
                
            # Condición para fechas
            elif pd.api.types.is_datetime64_any_dtype(datos_limpios):
                anys = datos_limpios.dt.year.value_counts().head(10)
                sns.barplot(x=anys.index, y=anys.values, color="forestgreen")
                plt.title(f"Análisis Temporal (Años): {columna}")
                
            # Condición por descarte (categóricas/texto)
            else:
                top_valores = datos_limpios.value_counts().head(10)
                sns.barplot(x=top_valores.values, y=top_valores.index, hue=top_valores.index, legend=False, palette="viridis")
                plt.title(f"Análisis Categórico (Top 10): {columna}")
                
            plt.tight_layout()
            # Guardamos especificando la ruta de la carpeta
            plt.savefig(f"mis_graficos/grafico_{columna}.png")
            plt.close() # importante cerrar la imagen porque sino se superpondrian unas a otras
            
        except Exception as e:
            print(f" Error inesperado en la columna '{columna}': {e}") #con el e mostramos el error especifico
            plt.close()
            
    print("\nProceso finalizado\nTodos los gráficos viables están guardados dentro de la carpeta 'mis_graficos'.")

#Cargamos el dataset 'titanic' de prueba
titanic = sns.load_dataset('titanic')

#Ejecutamos la función
generar_graficos_automaticos(titanic)
```

## 2.
Carrega l'arxiu matriu_distancies.xlsx a pandas, de manera que els noms de files i els noms de columnes siguin els de les ciutats. Borra "Las Palmas de Gran Canaria" i "Palma" perquè poguem fer el trajecte en cotxe.

Font: Mejores Rutas

Ens interessa visitar totes les ciutats principals d'Espanya recorrent la mínima distància possible.

No cal que ho facis de forma òptima, ens interessa que desenvolupis una solució raonable utilitzant les eines que tens actualment.

Per exemple, una aproximació senzilla (que no òptima) seria anant sempre a la ciutat més propera que no haguem visitat encara

Fes una funció que donada la matriu de distàncies i la ciutat d'origen, faci una proposta de ruta que sigui el més curta possible que puguis, retornant una llista amb l'ordre de visita. Dóna també la distància total recorreguda.

(EXTRA) Des de quina ciutat la ruta seria més curta amb l'algoritme plantejat

```python
import pandas as pd

# Cargamos el df usando la primera columna como los nombres de las filas ya q son las ciudades
df_rutas = pd.read_excel(r'C:\Users\aleja\Documents\Analisis de Datos\Especialidad IT ACADEMY\Sprint 10 Analítica de dades en Python\Sprint 10 ARCHIVOS\matriu_distancies.xlsx', index_col=0)

# Borramos las islas para poder viajar exclusivamente en coche
df_rutas = df_rutas.drop(index=["Palma", "Las Palmas de Gran Canaria"], 
                         columns=["Palma", "Las Palmas de Gran Canaria"])


def proponer_ruta(matriz, origen):
    # Lista de ciudades que nos faltan por visitar
    ciudades_por_visitar = list(matriz.index)
    ciudades_por_visitar.remove(origen)
    
    ruta = [origen] #creamos una lista para ir guardando las ciudades visitadas
    distancia_total = 0
    ciudad_actual = origen
    
    # El bucle se detendrá cuando la lista de pendientes se quede vacia
    while len(ciudades_por_visitar) > 0:
        # Filtramos la fila de la ciudad actual para ver solo las columnas pendientes
        distancias_opciones = matriz.loc[ciudad_actual, ciudades_por_visitar]
        
        # idxmin() nos da el nombre de la ciudad más cercana
        proxima_ciudad = distancias_opciones.idxmin()
        distancia_viaje = distancias_opciones.min()
        
        # Actualizamos las variables
        distancia_total += distancia_viaje
        ruta.append(proxima_ciudad)
        ciudades_por_visitar.remove(proxima_ciudad)
        
        # Saltamos a la nueva ciudad para la siguiente vuelta
        ciudad_actual = proxima_ciudad
        
    return ruta, distancia_total


#PROBAR CON UN ORIGEN (Ejemplo: Málaga) 
ruta_malaga, dist_malaga = proponer_ruta(df_rutas, "Málaga")
print("---  PRUEBA DE RUTA DESDE MÁLAGA ---")
print(f"Ruta propuesta: {ruta_malaga}")
print(f"Distancia total recorrida: {dist_malaga} km\n")


#RESOLVER EL (EXTRA)
# Creamos un diccionario para guardar cuánto se tarda empezando desde cada ciudad
registro_distancias = {}

for ciudad in df_rutas.index:
    #Desechamos la ruta con '_' y guardamos los kilómetros de ese viaje en 'kms'
    _, kms = proponer_ruta(df_rutas, ciudad) 
    # La clave (key) será el nombre de la ciudad y el valor (value) seran los kilómetros
    registro_distancias[ciudad] = kms

#Buscamos cuál es la ciudad que acumuló la menor distancia de todo el diccionario
mejor_origen = min(registro_distancias, key=registro_distancias.get) #con get buscamos el numero por kms y devuelve el nombre de la ciudad

print("EXTRA:")
print(f"La mejor ciudad para iniciar la ruta es: {mejor_origen}")
print(f"Consigue recorrer España entera haciendo solo: {registro_distancias[mejor_origen]} km")
```
