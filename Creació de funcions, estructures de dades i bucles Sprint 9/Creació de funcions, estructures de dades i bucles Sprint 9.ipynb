### 1. Calculadora de l'índex de massa corporal

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades. Pots obtenir més informació del seu càlcul en: 🔗 Índice de masa corporal IMC que es y como se calcula.

La funció ha de classificar el resultat en les seves respectives categories.

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduïdes per l’usuari/a no estan en el format adequat o no pren valors raonables.


```python
UMBRAL_BAJO_PESO = 18.5
UMBRAL_NORMAL = 25
UMBRAL_SOBREPESO = 30

def calcular_imc(peso,altura):
    imc=peso/(altura**2)
    
    if imc < UMBRAL_BAJO_PESO:
        return imc,"Bajo peso"
    
    elif imc < UMBRAL_NORMAL:
        return imc,"Peso normal"
    
    elif imc < UMBRAL_SOBREPESO:
        return imc,"Sobre peso"
    
    else:
        return imc,"Obesidad"

def pedir_numero(mensaje, minimo, maximo):
    while True:
        try:
            valor = float(input(mensaje).replace(",", "."))
        except ValueError:
            print("Introduce un número válido.")
            continue

        if not (minimo <= valor <= maximo):
            print(f"El valor debe estar entre {minimo} y {maximo}")
            continue

        return valor
 
def ejecutar_calculadora_IMC():   
    peso = pedir_numero("Introduce el peso (kg): ", 1, 500)
    altura = pedir_numero("Introduce la altura (m): ", 0.5, 3)

    imc,categoria=calcular_imc(peso,altura)
    print(f"El IMC es: {round(imc,2)} \nCategoria: {categoria} ")

if __name__ == "__main__":
    ejecutar_calculadora_IMC()
```

### 2. Convertidor de temperatures

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduides per l’usuari/a no estan en el format adequat.

(EXTRA): Pensa una manera d’emmagatzemar totes les possibles conversions en un sol objecte (Llista? Diccionari? DataFrame?) en comptes d’escriure molts if else en funció de la temperatura d’origen i la temperatura de destí.

```python
# Creé un diccionario anidado llamado conversiones. La primera clave representa la unidad de origen, 
# y dentro de cada una hay otro diccionario donde las claves son las unidades de destino. 
# Cada conversión se guarda como una función lambda, que recibe la temperatura introducida por el usuario y devuelve el resultado 
# convertido y redondeado por round

 #Creamos un diccionario para evitar el maximo de temperatura, el cero absoluto:
LIMITES = {
        "C": -273.15,
        "F": -459.67,
        "K": 0,
        "RA": 0,
        "RE": -218.52
}

conversiones={
    "C": {
        "C": lambda t: t,
        "F": lambda t: round((t * 9/5) + 32, 2),
        "K": lambda t: round(t + 273.15, 2),
        "RA": lambda t: round((t + 273.15) * 9/5, 2),
        "RE": lambda t: round(t * 0.8, 2)
    },
    
    "F": {
        "F": lambda t: t,
        "C": lambda t: round((t - 32) * 5/9, 2),
        "K": lambda t: round((t - 32) * 5/9 + 273.15, 2),
        "RA": lambda t: round(t + 459.67, 2),
        "RE": lambda t: round((t - 32) * 4/9, 2)
    },
    
    "K": {
        "K": lambda t: t,
        "C": lambda t: round(t - 273.15, 2),
        "F": lambda t: round((t - 273.15) * 9/5 + 32, 2),
        "RA": lambda t: round(t * 9/5, 2),
        "RE": lambda t: round((t - 273.15) * 0.8, 2)
    },
    
    "RA": {
        "RA": lambda t: t,
        "C": lambda t: round((t - 491.67) * 5/9, 2),
        "F": lambda t: round(t - 459.67, 2),
        "K": lambda t: round(t * 5/9, 2),
        "RE": lambda t: round((t - 491.67) * 4/9, 2)
    },
    
    "RE": {
        "RE": lambda t: t,
        "C": lambda t: round(t * 1.25, 2),
        "F": lambda t: round((t * 1.25) * 9/5 + 32, 2),
        "K": lambda t: round(t * 1.25 + 273.15, 2),
        "RA": lambda t: round((t * 1.25 + 273.15) * 9/5, 2)
    }
}

def pedir_datos():
    #aplicamos un while para repetir el bucle o las preguntas hasta que el usuario quiera salir o introduzca valores validos
    while True:
        unidad=input("Introduce la unidad de origen 'C', 'F', 'K', 'RA', 'RE' o SALIR para finalizar el programa : ").upper() #Aseguramos la mayuscula con upper
    
        if unidad=="SALIR":
            print("Programa Finalizado")
            return
    
        if unidad not in ('C', 'F', 'K', 'RA', 'RE'):
            print("Introduce una unidad válida")
            continue
    
        try:
            temperatura=float(input(f"Introduce la temperatura en grados °{unidad}: ").replace(",","."))
        
            if temperatura < LIMITES[unidad] :
                print("No es posible una temperatura menor al cero absoluto, ingresa una válida")
                continue
        
        except ValueError:
            print("Introduce una temperatura válida")
            
        return unidad,temperatura
            
def convertir_temperatura(unidad,temperatura):
    resultado = {}

    for destino, formula in conversiones[unidad].items():
        resultado[destino] = formula(temperatura)

    return resultado

def mostrar_conversiones(resultado):
    print("Conversiones:")

    for destino, valor in resultado.items():
        print(f"°{destino}: {valor}")
        
def ejecutar_conversor():
    datos = pedir_datos()

    if datos is None:
        return  # por si el usuario escribe SALIR

    unidad,temperatura=datos
    resultado=convertir_temperatura(unidad,temperatura)
    mostrar_conversiones(resultado)

    
if __name__ == "__main__":
    ejecutar_conversor()
```

### 4. Diccionari invers (amb possibilitat de duplicats)

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.

```python
#Esta funcion recibe un diccionario que previamente rellena el usuario, invierte las claves por los valores 
#Si hay duplicados mostrará un mensaje de advertencia y el diccionario completo
#sino mostrará un mensaje de que no hay valores repetidos y mostrará el diccionario completo

def diccionario_inverso(diccionario):
    
    diccionario_invertido={}
    
    for clave, valor in diccionario.items():
        
        if valor not in diccionario_invertido:
            diccionario_invertido[valor]=clave
             
        else: 
                                                            
            if isinstance(diccionario_invertido[valor], list): #comprobamos con la funcion isinstance si ya tiene asociada una lista de claves repetidas. Si es así, añadimos la nueva clave usando append().
                 diccionario_invertido[valor].append(clave)    # la condicion tambien podria hacerse con: type(diccionario_invertido[valor]) == list
            
            else:
                diccionario_invertido[valor]=[diccionario_invertido[valor],clave]
        
    return diccionario_invertido

diccionario={}

for i in range (3):
    clave=input(f"Introduce la clave {i+1}: ")
    clave=clave.capitalize()
    
    valor=input(f"Introduce el valor de la {clave}: ")
    valor=valor.capitalize()
    
    diccionario[clave]=valor


resultado=diccionario_inverso(diccionario)

duplicados=False

for clave,valor in resultado.items():
    
    if isinstance(resultado[clave],list):  #comprobamos si es una lista para saber que hay duplicados
        duplicados=True
        
    
if duplicados:
    print("Se han detectado valores repetidos en la encuesta")
    
    if isinstance(valor, list): #
        print(f"La clave {clave} está asociada a estos valores: {valor}")
            
else:
    print("No se han detectado valores repetidos.")

print("\nDiccionario invertido completo:")
for clave, valor in resultado.items():
    print(f"{clave}: {valor}")
        
```

### 1. Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

```python
#Aplicamos with open para abrir el archivo del ejercicio 

def leer_archivo(ruta):
    with open(ruta,"r",encoding="utf-8") as archivo:
        return archivo.read()

def limpieza_texto(texto):
    texto=texto.lower() #evitamos mayusculas 
    texto_limpio="" 
    
    for caracter in texto: #iteramos el texto para limpiarlo y quedarnos solo con letras del abecedario 
        if caracter.isalpha():
            texto_limpio+=caracter
    
        elif caracter.isspace(): #tambien conservamos los espacios
            texto_limpio+=caracter
        
    texto_limpio=texto_limpio.split() #separamos el texto en palabras y convertimos a su vez en lista 
    return texto_limpio

def contar_palabras(texto_limpio):
    diccionario_poema={}

    for palabra in texto_limpio: #iteramos ahora el texto limpio para crear el diccionario con clave como la primera letra de cada palabra 
        primera_letra=palabra[0]
    
        if primera_letra not in diccionario_poema: 
            diccionario_poema[primera_letra]={}
    
        if palabra not in diccionario_poema[primera_letra]:
            diccionario_poema[primera_letra][palabra]=1
        
        else:
            diccionario_poema[primera_letra][palabra]+=1
    return diccionario_poema
            
#ahora copiamos todo pero de forma ordenada en un diccionario nuevo
def dict_ordenado(diccionario_poema):
    diccionario_ordenado={}
    
    for clave in sorted(diccionario_poema):
        diccionario_ordenado[clave]={} # Creo la letra ordenada, pero vacia y luego voy metiendo dentro sus palabras ordenadas.
    
        for palabra in sorted(diccionario_poema[clave]):
        
            diccionario_ordenado[clave][palabra]=diccionario_poema[clave][palabra]
    return diccionario_ordenado

def mostrar_resultado(diccionario_ordenado):
#forma de imprimir el diccionario anidado
    print("Resultado ordenado: \n")

    for clave_principal in diccionario_ordenado:
        print(f"{clave_principal.upper()}")
    
        for palabra,contador in diccionario_ordenado[clave_principal].items():
            print(f"     {palabra}: {contador}")
            
def ejecutar_programa_dict():
    texto=leer_archivo("tu me quieres blanca.txt") #El nombre del archivo, ya q la ruta sera la misma carpeta 
    texto_limpio=limpieza_texto(texto)
    diccionario_poema=contar_palabras(texto_limpio)
    diccionario_ordenado=dict_ordenado(diccionario_poema)
    mostrar_resultado(diccionario_ordenado)

if __name__ == "__main__":
    ejecutar_programa_dict()
```

### 2. Conversió de tipus de dades.

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:

```python

datos = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']] # el 1e10 python lo toma como notacion cientifica y lo convierte en float

def conversiones(datos):
    convertibles=[]
    no_convertibles=[]

    for elemento in datos:
    
        if isinstance(elemento,tuple) or isinstance(elemento,list): #mas limpio seria: if isinstance(elemento, (tuple, list)):
        
            for subelemento in elemento:
                try:
                    convertibles.append(float(subelemento))
        
                except:
                    no_convertibles.append(subelemento)
        else:
            try:
                convertibles.append(float(elemento))
        
            except:
                no_convertibles.append(elemento)
        
    return convertibles,no_convertibles
        
convertibles,no_convertibles=conversiones(datos)
print(convertibles)
print(no_convertibles)

```

```python

#Creamos los diccionarios necesarios que llenaremos de iterar el archivo
diccionario_goles={}
diccionario_goles_contra={}
diccionario_puntos={}

with open(r"C:\Users\aleja\Documents\Analisis de Datos\Especialidad IT ACADEMY\Sprint 9 Nocions i coneixements bàsics de Python\Archivos Sprint 9\historic partits.txt", "r", encoding="utf-8") as archivo:

    for linea in archivo:
        
        linea=linea.strip() #limpiamos la linea, elimina el \n al final 
        
        #convertimos en lista y separamos los elementos por el tabulador q serian los espacios 
        local,resultado,visitante=linea.split("\t") 
        #ahora con el resultado obtenemos los goles del equipo local y del visitante
        goles_local, goles_visitante = resultado.split("-") #si ponemos por posiciones no seria seguro, mejor separar con el guion
        goles_local=int(goles_local)
        goles_visitante=int(goles_visitante)
        
        #rellenando el diccionario goles 
        if local not in diccionario_goles:
            diccionario_goles[local]=goles_local
        else:
            diccionario_goles[local]+=goles_local  #si ya esta sumamos los goles
        
        if visitante not in diccionario_goles:
            diccionario_goles[visitante]=goles_visitante
        else:
            diccionario_goles[visitante]+=goles_visitante
                 
        #rellenando el diccionario goles en CONTRA         
        if local not in diccionario_goles_contra:
            diccionario_goles_contra[local]=goles_visitante
        else:
            diccionario_goles_contra[local]+=goles_visitante
        
        if visitante not in diccionario_goles_contra:
            diccionario_goles_contra[visitante]=goles_local
        else:
            diccionario_goles_contra[visitante]+=goles_local
            
        #Rellenando el diccionario de puntos
       # nos aseguramos de que ambos equipos existen en el diccionario creandolos con 0 pts
        if local not in diccionario_puntos:
            diccionario_puntos[local] = 0

        if visitante not in diccionario_puntos:
            diccionario_puntos[visitante] = 0
            
        #luego agregamos los puntos segun los resultados del partido:
        if goles_local > goles_visitante:
            diccionario_puntos[local] += 3

        elif goles_local < goles_visitante:
            diccionario_puntos[visitante] += 3

        else:
            diccionario_puntos[local] += 1
            diccionario_puntos[visitante] += 1
        
#Buscamos el equipo mas goleador
maximo_goles=max(diccionario_goles.values())#buscamos el valor maximo de goles
for equipo,goles in diccionario_goles.items(): #iteramos el diccionario en pares y comparamos con la variable y ahi imprimimos el nombre y numero
    if goles==maximo_goles:
        equipo_goleador,goles=equipo,goles
            
            
#Buscamos el equipo que ha recibido mas goles, el mas goleado:
maximo_goles_contra=max(diccionario_goles_contra.values())
for equipo,goles in diccionario_goles_contra.items(): 
    if goles==maximo_goles_contra:
        equipo_goleado,goles_contra=equipo,goles

#ordenamos los equipos por puntos para tener la clasificacion global:
    
#ordenamos usando key= donde lambda recibe los puntos y ordena por los puntos que estan en la posicion 1, y reverse para ordenar de mayor a menor
clasificacion_global=sorted(diccionario_puntos.items(),key= lambda puntos: puntos[1],reverse=True) 
            
#El nombre total de gols que ha fet cada equip.
goles_por_equipo=sorted(diccionario_goles.items(),key= lambda goles:goles[1],reverse=True)
    
print("\nGOLES POR EQUIPO")
print("-" * 40)
print(f"{'Pos':<5}{'Equipo':<25}{'Goles':>8}")
print("-" * 40)

for posicion, (equipo, goles) in enumerate(goles_por_equipo, start=1):
    print(f"{posicion:<5}{equipo:<25}{goles:>8}")
        
#El nom de l’equip més golejador.
print("\nEl equipo mas goleador: \n")
print(f"{equipo_goleador} : {maximo_goles} ")

#El nom de l’equip més golejat
print("\nEl equipo con mas goles en contra: \n")
print(f"{equipo_goleado}: {goles_contra}\n")

#La classificació global (cada victòria: 3 pts, empat 1 pts, derrota 0 pts)
print("\nCLASIFICACIÓN GLOBAL")
print("-" * 40)
print(f"{'Pos':<5}{'Equipo':<25}{'Puntos':>8}")
print("-" * 40)

for posicion, (equipo, puntos) in enumerate(clasificacion_global, start=1):
    print(f"{posicion:<5}{equipo:<25}{puntos:>8}")
```
